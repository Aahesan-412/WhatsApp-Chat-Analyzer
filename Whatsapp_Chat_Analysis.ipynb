### WhatsApp Chat Analysis

```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with TY BCA.....🏢.txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(type(data))
```

```python
print(data)
```

```python
###  times and  dates are remove drom our data

pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}[\s\u202f](?:AM|PM)\s-\s'
```

```python
message = re.split(pattern, data)[1:]
message
```

```python
### Dates Seperated

dates = re.findall(pattern,data)
dates = [d.replace('\u202f','')for d in dates]
dates
```

```python
## Create Dataframe

df = pd.DataFrame({'user_message':message, 'message_date': dates})

## Covert message data types
df['message_date'] = pd.to_datetime(df['message_date'], format='%m/%d/%y, %I:%M%p - ')


df.rename(columns={'message_date': 'date'}, inplace=True)

df.head()
```

```python
df.shape
```

```python
print(df.columns)
```

```python
## Seperate users and message

users = []
messages = []

for message in df['user_message']:
    # Split only on the FIRST occurrence of ': ' 
    entry = re.split('([\w\W]+?):\s', message)
    
    if len(entry) > 1: # If a colon was found, it's a user message
        users.append(entry[1])      # User Name
        messages.append(entry[2])    # The actual message (Fixed: added 's' to messages)
    else:
        users.append("group_notification")
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)

df.head()
```

```python
df['year'] = df['date'].dt.year  ## Year are seperated
```

```python
df['month'] = df['date'].dt.month_name() ## Month sepearated
```

```python
df['day'] = df['date'].dt.day    ## day seperated
```

```python
df['hour'] = df['date'].dt.hour  ## hour seperated
```

```python
df['minute'] = df['date'].dt.minute  ## minute Seperated
```

```python
##remove  group notification

temp = df[df['user'] != 'group_notification']

## Remove Media ommited 

temp = temp[temp['message'] != '<Media omitted>\n']
```

```python
## Remove Stop Words

f = open("stop_hinglish.txt",'r')  
stop_words = f.read()
```

```python
words = []

for message in temp['message']:
    for word in message.lower().split():
        if word not in stop_words:
            words.append(word)
    
```

```python
from collections import Counter
import pandas as pd
pd.DataFrame(Counter(words).most_common(20), columns=['Word','No of Times'])
```

```python
import emoji

emojis = []

for message in df['message']:
    emojis.extend([c for c in message if c in emoji.EMOJI_DATA])
    
emoji_df = pd.DataFrame(Counter(emojis).most_common(len(Counter(emojis))),columns=['Emojis','No of Times'])
```

```python
## most Avtivity on which day 

df['day_name'] = df['date'].dt.day_name()
```

```python
df['day_name'].value_counts()
```

```python
## Heatmap which time user more active
period = []
for hour in df[['day_name','hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str('00'))
    elif hour == 0:
        period.append(str('00')+ "-" + str(hour+1))
    else:
        period.append(str(hour) + "-" + str(hour+1))
```

```python
df['period'] = period
```

```python
import seaborn as sns
plt.figure(figsize=(20,6))
sns.heatmap(df.pivot_table(index='day_name', columns='period', values='message', aggfunc='count').fillna(0))
plt.yticks(rotation='horizontal')
plt.show()
```
